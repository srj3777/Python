```python

import pandas as pd
import numpy as np

# ------------------------------------------------------------
# 1. Load CSV and Saving to CSV
# ------------------------------------------------------------
df = pd.read_csv("/content/drive/MyDrive/LAB/random_marks.csv")
top3=df.head(3)
top3.to_csv('/content/drive/MyDrive/LAB/top3.csv', index=False)


# ------------------------------------------------------------
# 2. Explore Data
# ------------------------------------------------------------
display(df.head())
display(df.tail())
df.info()
df.describe()
df.shape
df.columns
df.dtypes
df["NEWLANG"]=df["JAVA"]+"-"+df["C"]
print(df)

# ________________________________________
# ------------------------------------------------------------
# 3. Clean Special Characters in Marks
# ------------------------------------------------------------
mark_columns = ["Python", "JAVA", "NET", "C", "ANGULAR"]

for col in mark_columns:
    df[col] = df[col].astype(str).str.replace(r'[^0-9]', '', regex=True)
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.head()
# ________________________________________
# ------------------------------------------------------------
# 4. Handle Missing Values
# ------------------------------------------------------------
df.isna().sum()
df.fillna(0, inplace=True)
df.head()
# ________________________________________
# ------------------------------------------------------------
# 5. Add Total & Average Columns
# ------------------------------------------------------------
df["Total"] = df[mark_columns].sum(axis=1)
df["Average"] = df[mark_columns].mean(axis=1)

df.head()
# ________________________________________
# ------------------------------------------------------------
# 6. Filtering Examples
# ------------------------------------------------------------
df[df["Python"] > 50]
df[df["JAVA"] == 0]
df[(df["NET"] > 40) & (df["ANGULAR"] < 60)]
df.query("C > 70")
# ________________________________________
# ------------------------------------------------------------
# 7. Sorting Examples
# ------------------------------------------------------------
df.sort_values(by="Total", ascending=False).head()
df.sort_values(by=["Python", "JAVA"], ascending=[False, True]).head()
# ________________________________________
# ------------------------------------------------------------
# 8. NumPy Operations
# ------------------------------------------------------------
arr = df[mark_columns].to_numpy()

mean_vals = np.mean(arr, axis=0)
max_vals = np.max(arr, axis=0)
min_vals = np.min(arr, axis=0)
std_vals = np.std(arr, axis=0)

mean_vals, max_vals, min_vals, std_vals
# ________________________________________
# ------------------------------------------------------------
# 9. NumPy Filtering
# ------------------------------------------------------------
high_python = arr[arr[:, 0] > 70]
high_python
# ________________________________________
# ------------------------------------------------------------
# 10. Save Cleaned CSV
# ------------------------------------------------------------
df.to_csv("cleaned_marks.csv", index=False)
print("Cleaned CSV saved as cleaned_marks.csv")
```
